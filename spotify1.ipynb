```python
!pip3 install scikit-learn
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import MinMaxScaler
```

```python


# Fetch the "Rap Caviar" playlist ID
def fetch_rap_caviar_playlist_id(spotify):
    rap_caviar_playlist_id = None
    results = spotify.search(q='Rap Caviar', type='playlist', limit=10)
    for playlist in results['playlists']['items']:
        if playlist['name'] == 'RapCaviar':
            rap_caviar_playlist_id = playlist['id']
            break
    return rap_caviar_playlist_id

# Fetch the tracks and audio features of the "Rap Caviar" playlist
def fetch_rap_caviar_tracks(spotify, playlist_id):
    tracks = []
    results = spotify.playlist_tracks(playlist_id)
    for track_data in results['items']:
        track = track_data['track']
        audio_features = spotify.audio_features(track['id'])[0]

        track_info = {
            'id': track['id'],
            'name': track['name'],
            'artist': track['artists'][0]['name']
        }
        track_info.update(audio_features)
        tracks.append(track_info)
    return tracks

# Fetch the "Rap Caviar" playlist ID
rap_caviar_playlist_id = fetch_rap_caviar_playlist_id(spotify)

# Fetch the tracks and audio features of the "Rap Caviar" playlist
rap_caviar_tracks = fetch_rap_caviar_tracks(spotify, rap_caviar_playlist_id)

# Create a DataFrame with the collected data
df = pd.DataFrame(rap_caviar_tracks)

# Now, you can use the kNN model provided in the previous response to build a recommender based on the "Rap Caviar" playlist.
```

```python



# Fetch the "Rap Caviar" playlist ID and tracks
rap_caviar_playlist_id = fetch_rap_caviar_playlist_id(spotify)
rap_caviar_tracks = fetch_rap_caviar_tracks(spotify, rap_caviar_playlist_id)

# Create a DataFrame with the collected data
df = pd.DataFrame(rap_caviar_tracks)

# Select the audio features you want to use for building the recommendation model
audio_features = [
    'danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
    'instrumentalness', 'liveness', 'valence', 'tempo'
]

# Scale the features to bring them to a common scale
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df[audio_features])

# Train a kNN model
knn = NearestNeighbors(n_neighbors=10, metric='cosine')
knn.fit(scaled_features)

# Function to recommend songs based on the kNN model
def recommend_songs(song_id, knn_model, df, num_recommendations=10):
    song_index = df[df['id'] == song_id].index[0]
    distances, indices = knn_model.kneighbors(scaled_features[song_index].reshape(1, -1))
    
    recommended_song_ids = []
    for idx in indices.flatten():
        if idx != song_index:
            recommended_song_ids.append(df.iloc[idx]['id'])
    
    return recommended_song_ids[:num_recommendations]

# Example usage: recommend songs similar to a given song ID from the "Rap Caviar" playlist
input_song_id = '7aRCf5cLOFN1U7kvtChY1G'  # Replace this with a song ID from the "Rap Caviar" playlist
recommended_songs = recommend_songs(input_song_id, knn, df)

def print_song_details(song_ids, df):
    for song_id in song_ids:
        song = df[df['id'] == song_id].iloc[0]
        print(f"Song ID: {song['id']}\nSong Name: {song['name']}\nArtist: {song['artist']}\n")

# Example usage:
print("Recommended songs:")
print_song_details(recommended_songs, df)

```
