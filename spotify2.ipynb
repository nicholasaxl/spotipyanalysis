```python
# get all the featured playlists
featured_playlists = spotify.featured_playlists(limit=None)

# loop through the featured playlists and print their names
for playlist in featured_playlists['playlists']['items']:
    print(playlist['name'])

# get all the genre-based playlists
genres = spotify.categories()['categories']['items']
for genre in genres:
    print(genre['name'])
    playlists = spotify.category_playlists(category_id=genre['id'], limit=None)
    for playlist in playlists['playlists']['items']:
        if not playlist:
            print(f"ERROR: {genre['name']} does not have a playlist")
            continue
        print('\t' + playlist['name'])
```
