```python
for playlist in playlists['items']:
    print(playlist['name'])
    print('  total tracks', playlist['tracks']['total'])
    
    # Get track IDs of the playlist
    results = spotify.playlist(playlist['id'], fields="tracks,next")
    tracks = results["tracks"]
    track_ids = [track["track"]["id"] for track in tracks["items"]]
    while tracks["next"]:
        tracks = spotify.next(tracks)
        track_ids += [track["track"]["id"] for track in tracks["items"]]
    
    # Get audio features of each track in the playlist
    features = []
    for i in range(0, len(track_ids), 50):
        audio_features = spotify.audio_features(track_ids[i:i+50])
        for track in audio_features:
            features.append(track)
    
    # Create a vector of feature values
    feature_names = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']
    feature_vector = []
    for name in feature_names:
        feature_values = [track[name] for track in features]
        feature_vector.append(feature_values)
    
    # Print the feature vector
    print("Feature vector:")
    for i, name in enumerate(feature_names):
        print(name, feature_vector[i])
```
